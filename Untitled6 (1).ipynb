```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms


class PrunableLinear(nn.Module):
    def __init__(self, in_features, out_features, temperature=1.0):
        super().__init__()

        self.weight = nn.Parameter(torch.randn(out_features, in_features) * 0.02)
        self.bias = nn.Parameter(torch.zeros(out_features))


        self.gate_scores = nn.Parameter(torch.randn(out_features, in_features) - 1.0)

        self.temperature = temperature

    def forward(self, x):
        gates = torch.sigmoid(self.gate_scores / self.temperature)
        return F.linear(x, self.weight * gates, self.bias)

    def get_gates(self):
        return torch.sigmoid(self.gate_scores / self.temperature)

    def set_temperature(self, temp):
        self.temperature = temp


class PrunableNet(nn.Module):
    def __init__(self):
        super().__init__()

        self.conv = nn.Sequential(
            nn.Conv2d(3, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),

            nn.Conv2d(64, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),

            nn.Conv2d(128, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )

        self.fc1 = PrunableLinear(128 * 8 * 8, 256)
        self.dropout = nn.Dropout(0.5)
        self.fc2 = PrunableLinear(256, 10)

    def forward(self, x):
        x = self.conv(x)
        x = x.view(x.size(0), -1)
        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x


def compute_sparsity_loss(model):
    total, count = 0.0, 0

    for m in model.modules():
        if hasattr(m, "get_gates"):
            g = m.get_gates()


            total += torch.sum(g)

            count += g.numel()

    return total / count


def compute_current_sparsity(model, threshold=1e-2):
    total, zero = 0, 0

    for m in model.modules():
        if hasattr(m, "get_gates"):
            g = m.get_gates()
            total += g.numel()
            zero += (g < threshold).sum().item()

    return zero / total



def get_lambda(epoch, total_epochs, base_lambda=1e-2):
    progress = epoch / total_epochs


    return base_lambda * (progress ** 2)


device = "cuda" if torch.cuda.is_available() else "cpu"



transform_train = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform_train)

testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform_test)

trainloader = torch.utils.data.DataLoader(trainset, batch_size=128, shuffle=True)
testloader = torch.utils.data.DataLoader(testset, batch_size=128, shuffle=False)


def train(model, epochs=40, base_lambda=1e-2):
    model.to(device)

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-3)

    scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=15, gamma=0.5)

    for epoch in range(epochs):
        model.train()


        if epoch < 10:
            lambda_sparse = 0
        else:
            lambda_sparse = get_lambda(epoch, epochs, base_lambda)


        temp = max(0.1, 1.0 - epoch / epochs)

        for m in model.modules():
            if hasattr(m, "set_temperature"):
                m.set_temperature(temp)

        for x, y in trainloader:
            x, y = x.to(device), y.to(device)

            optimizer.zero_grad()

            out = model(x)
            cls_loss = criterion(out, y)
            sp_loss = compute_sparsity_loss(model)

            loss = cls_loss + lambda_sparse * sp_loss

            loss.backward()
            optimizer.step()

        scheduler.step()

        sparsity = compute_current_sparsity(model)

        print(f"Epoch {epoch+1}: λ={lambda_sparse:.5f}, Sparsity={sparsity:.2f}")


def evaluate(model):
    model.eval()
    correct, total = 0, 0

    with torch.no_grad():
        for x, y in testloader:
            x, y = x.to(device), y.to(device)
            out = model(x)
            _, pred = torch.max(out, 1)
            total += y.size(0)
            correct += (pred == y).sum().item()

    return 100 * correct / total, compute_current_sparsity(model)


def run():
    lambdas = [0.005, 0.01, 0.02]

    for lam in lambdas:
        print(f"\nLambda: {lam}")

        model = PrunableNet()
        train(model, epochs=40, base_lambda=lam)

        acc, sp = evaluate(model)

        print(f"Accuracy: {acc:.2f}%, Sparsity: {sp:.2f}")


if __name__ == "__main__":
    run()
```
